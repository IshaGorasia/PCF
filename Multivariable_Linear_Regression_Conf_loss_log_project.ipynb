```python
dataset = pd.read_excel("pcf_modeSoln_data_1.xlsx", sheet_name="SiO2-air-rings-5-dBYp-0.6")
x = dataset.iloc[:, 1:6].values
y = dataset.iloc[:, 10].values
for i in x:
  X.append(i)
for i in y:
  Y.append(i)

print(len(X))
print(len(Y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data_X = scaler.fit_transform(X)
```

```python
Y = Y.reshape(-1,1)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data_Y = scaler.fit_transform(Y)
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100)
```

```python
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print("r2 score is: ",score)
print("mean_absolute_error is: " ,mean_absolute_error(y_test,y_prediction)) 
print("mean_squared_error is: " ,mean_squared_error(y_test,y_prediction))
print("root_mean_squared error of is: ",np.sqrt(mean_squared_error(y_test,y_prediction)))
```
